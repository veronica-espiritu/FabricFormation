```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit # 'col' para referenciar columnas, 'lit' para valores constantes
```

```python
spark = SparkSession.builder.getOrCreate() # Se crea una sesión de spark
```

```python
df_sales = spark.read.format("delta").load("Tables/sales") # Se lee la tabla sales
```

```python
df_rates = spark.read.format("delta").load("Tables/exchange_rates") # Se lee la tabla exchange_rates
```

```python
rates_row = df_rates.collect()[0].asDict() # Se extrae la fila como un diccionario para acceder a los valores directamente
```

```python
currency = "USD" # Moneda de destino
column_name = f"rates.{currency}" # Nombre de la columna que contiene el tipo de cambio
```

```python
exchange_rate = rates_row[column_name] # Se obtiene el valor del tipo de cambio de la columna
```

```python
df_sales_converted = df_sales.withColumn(
    f"price_{currency}",        # Nombre de la columna 
    col("quantity") * lit(exchange_rate) # Se multiplica cada valor de 'quantity' por el tipo de cambio ('USD')
    ) # 
```

```python
df_sales_converted.write.format("delta").mode("overwrite").saveAsTable("silver_lakehouse.sales_converted")
```
